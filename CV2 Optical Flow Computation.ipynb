```python
def countdir(folder_path):
    return len(glob.glob(f'{folder_path}/*'))
```

```python
def isFlowCorrect(x, y, threshold=1e-9):
    return not(np.isnan(x) or np.isnan(y) or abs(x) < threshold or abs(y) < threshold)
```

```python
def normalize_flow_pixel(v, L, H):
    if v < L or L == H:
        return np.uint8(0)
    elif v > H:
        return np.uint8(255)
    else:
        pixel = 255.0 * ((v  - L) / (H - L))
        pixel = pixel + 0.50 if pixel > 0 else pixel - 0.50
        return pixel
    
def convertFlowToImageOriginal(flow, L, H):
    flow_n = np.vectorize(lambda v: normalize_flow_pixel(v, L, H))(flow).astype(np.uint8)
    
    return flow_n
```

```python
def convertFlowToImage(flow, L, H):
    if L == H:
        return np.zeros(shape=flow.shape, dtype=np.uint8)
    else:
        flow_n = 255 * (flow - L) / (H - L)
        flow_n = np.round(flow_n).astype(np.uint8)
        return flow_n
```

```python
def try_makedirs(file_path):
    try:
        os.makedirs(file_path)
    except:
        pass
```

```python
def is_created(destination_folder, folder):
    # Optical Flow Frames Stats
    stats_created = os.path.exists(f'{destination_folder}/flow/stats/{folder}.pkl')
    
    # get number of frames
    rgb_frames = countdir(f'{destination_folder}/rgb/{folder}')
    flow_x_frames = countdir(f'{destination_folder}/flow/x/{folder}')
    flow_y_frames = countdir(f'{destination_folder}/flow/y/{folder}')
    # check if rgb frames - 1 (first frame has no previous frame) equals amount of optical flow frames
    of_created = (rgb_frames - 1) == flow_x_frames and (rgb_frames - 1) == flow_y_frames
    
    return of_created, stats_created
```

```python
def update_of_stats(of_stats, FlowXN, FlowYN):
    # Optical Flow Stats X
    of_stats['x']['mins'].append(FlowXN.min())
    of_stats['x']['maxs'].append(FlowXN.max())
    of_stats['x']['means'].append(FlowXN.mean())
    of_stats['x']['medians'].append(np.median(FlowXN))
    of_stats['x']['stds'].append(FlowXN.std())
    # Optical Flow Stats Y
    of_stats['y']['mins'].append(FlowYN.min())
    of_stats['y']['maxs'].append(FlowYN.max())
    of_stats['y']['means'].append(FlowYN.mean())
    of_stats['y']['medians'].append(np.median(FlowYN))
    of_stats['y']['stds'].append(FlowYN.std())
```

# Optical Flow

```python
def main(dataset, debug=False, folder_names=None, start_frame=None, stop_percentage=None):
    DATASET_FOLDER = f'C:/Users/markw/Downloads/{dataset}/rgb'
    DESTINATION_FOLDER = f'C:/Users/markw/Downloads/{dataset}'
    OF = cv2.cuda.BroxOpticalFlow_create(0.197, 50.0, 0.8, 10, 77, 10)
    DEST_MAT = cv2.cuda_GpuMat()
    JPEG_QUALITY = 95
    
    if folder_names == None:
        folder_names = [f.split('\\')[-1] for f in glob.glob(f'{DATASET_FOLDER}/*')]
        list.sort(folder_names)
        
    print(f'Found {len(folder_names)} folfers')
    
    for idx, folder in tqdm(enumerate(folder_names), total=len(folder_names)):
        # stop when percentage
        if stop_percentage and idx / len(folder_names) * 100 > stop_percentage:
            return
        
        
        # check if already created, next folder should exist
        of_created, stats_created = is_created(DESTINATION_FOLDER, folder)
        
        if debug:
            print(f'of_created: {of_created}, stats_created: {stats_created}')
            
        if of_created and stats_created:
            continue
        
        # Result Folders
        base_dir = f'{DATASET_FOLDER}/{folder}'
        folder_x = f'{DESTINATION_FOLDER}/flow/x/{folder}'
        folder_y = f'{DESTINATION_FOLDER}/flow/y/{folder}'
        try_makedirs(folder_x)
        try_makedirs(folder_y)
        n_frames = countdir(base_dir)
        of_stats = dict({
            'x': dict({ 'mins': [], 'maxs': [], 'means': [], 'medians': [], 'stds': [], 'sums': [], }),
            'y': dict({ 'mins': [], 'maxs': [], 'means': [], 'medians': [], 'stds': [], 'sums': [], }),
        })
        
        for frame_idx in range(start_frame if start_frame else 2, n_frames + 1):
            # Read previous and current frame to compute optical flow
            file_name_prev = '%05d.jpg' % (frame_idx-1)
            file_name_cur = '%05d.jpg' % frame_idx
            file_path_prev = f'{base_dir}/{file_name_prev}'
            file_path_cur = f'{base_dir}/{file_name_cur}'
            
            frame_prev = cv2.imread(file_path_prev, cv2.IMREAD_GRAYSCALE)
            frame_cur  = cv2.imread(file_path_cur, cv2.IMREAD_GRAYSCALE)
            
            if debug:
                print(f'n_frames: {n_frames}')
                print(f'file_path_prev: {file_path_prev}, min: {frame_prev.min()}, max: {frame_prev.max()}')
                print(f'file_path_cur: {file_path_cur}, min: {frame_cur.min()}, max: {frame_cur.max()}')
            
            if frame_prev is None or frame_cur is None:
                print('Frames could not be read:')
                
            if debug:
                fig, axes = plt.subplots(1, 2, figsize=(12,5))
                axes[0].imshow(frame_prev)
                axes[1].imshow(frame_cur)
                plt.show()
        
            # Create Optical Flow Frame
            frame_prev_cuda = cv2.cuda_GpuMat(frame_prev.astype(np.float32) / 255)
            frame_cur_cuda = cv2.cuda_GpuMat(frame_cur.astype(np.float32) / 255)

            Flows = OF.calc(frame_prev_cuda, frame_cur_cuda, DEST_MAT).download()
            FlowX, FlowY = np.array_split(Flows, 2, axis=2)
            
            # Update Optical Flow Stats Before Normalization
            update_of_stats(of_stats, FlowX, FlowY)
            
            if debug:
                print(f'of_stats: {of_stats}')
            
            if debug:
                fig, axes = plt.subplots(1, 2, figsize=(12,5))
                axes[0].imshow(FlowX)
                axes[0].set_title(f'FlowX min: %.2f, max: %.2f' % (FlowX.min(), FlowX.max()))
                axes[1].imshow(FlowY)
                axes[1].set_title(f'FlowY min: %.2f, max: %.2f' % (FlowY.min(), FlowY.max()))
                plt.show()
            
            # Normalize Optical Flow Frame to UINT8
            min_x, max_x = FlowX.min(), FlowX.max()
            min_y, max_y = FlowY.min(), FlowY.max()
            
            MAX_ABS_X = max(abs(min_x), abs(max_x))
            MAX_ABS_Y = max(abs(min_y), abs(max_y))
            MAX_ABS_VAL = max(MAX_ABS_X, MAX_ABS_Y)
            
            min_x_f, min_y_f = -MAX_ABS_VAL, -MAX_ABS_VAL
            max_x_f, max_y_f = MAX_ABS_VAL, MAX_ABS_VAL
    
            FlowXN = convertFlowToImage(FlowX, min_x_f, max_x_f)
            FlowYN = convertFlowToImage(FlowY, min_y_f, max_y_f)
            
            if debug:
                print(f'min_x: %.2f, max_x: %.2f, min_y: %.2f, max_y: %.2f' % (min_x, max_x, min_y, max_y))
                print(f'MAX_ABS_X: %.2f, MAX_ABS_Y: %.2f, MAX_ABS_VAL: %.2f' % (MAX_ABS_X, MAX_ABS_Y, MAX_ABS_VAL))
                
                fig, axes = plt.subplots(1, 2, figsize=(12,5))
                axes[0].imshow(FlowXN)
                axes[0].set_title(f'FlowXN min: %.2f, max: %.2f' % (FlowXN.min(), FlowXN.max()))
                axes[1].imshow(FlowYN)
                axes[1].set_title(f'FlowYN min: %.2f, max: %.2f' % (FlowYN.min(), FlowYN.max()))
                plt.show()
            
            # write image
            file_path_x = f'{folder_x}/{file_name_prev}'
            file_path_y = f'{folder_y}/{file_name_prev}'
            
            if debug:
                print(f'file_path_x: {file_path_x}')
                print(f'file_path_y: {file_path_y}')
            else:
                if not of_created:
                    succes_u = cv2.imwrite(file_path_x, FlowXN) 
                    succes_v = cv2.imwrite(file_path_y, FlowYN) 
                    # check if images are succesfully written
                    if not(succes_u) or not(succes_v):
                        print(f'images not succesfully written to disk, folder: {folder}, frame_idx: {frame_idx}')
                
            if debug or start_frame:
                return frame_prev_cuda, frame_cur_cuda, DEST_MAT
        
        # save Optical Flow Frame Stats
        with bz2.BZ2File(f'{DESTINATION_FOLDER}/flow/stats/{folder}.pkl', 'wb') as f:
            cPickle.dump(of_stats, f)
```

```python
def get_vid(file_path, frame_idx):
    vid = cv2.VideoCapture(file_path)
    idx = 0
    while vid.isOpened():
        ret, frame = vid.read()
        if not ret:
            print('Frame index too high')
            return
        
        if ret and idx is frame_idx:
            return frame
        else:
            idx += 1

def get_img(file_path):
    return cv2.imread(file_path)
    
frame_0 = get_vid('./temp/videos/BOER.mp4', 58)
frame_0 = cv2.cvtColor(frame_0, cv2.COLOR_BGR2GRAY) / 255.0

frame_1 = get_vid('./temp/videos/BOER.mp4', 59)
frame_1 = cv2.cvtColor(frame_1, cv2.COLOR_BGR2GRAY) / 255.0

h, w = frame_0.shape
frame_0 = frame_0[:, (w-h)//2:(w-h)//2+h]
frame_1 = frame_1[:, (w-h)//2:(w-h)//2+h]

rows, cols = 1, 2
fig, axes = plt.subplots(nrows=rows, ncols=cols, figsize=(cols*6, rows*6))
axes[0].imshow(frame_0)
axes[0].set_title(f'shape: {frame_0.shape}, mean: {round(frame_0.mean(), 2)}, std: {round(frame_0.std(), 2)}')
axes[1].imshow(frame_1)
axes[1].set_title(f'shape: {frame_1.shape}, mean: {round(frame_1.mean(), 2)}, std: {round(frame_1.std(), 2)}')
plt.show()
```
