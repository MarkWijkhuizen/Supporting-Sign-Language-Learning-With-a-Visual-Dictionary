```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from tqdm.notebook import tqdm
from multiprocessing import cpu_count

import os
import io
import math
import sys
import glob
import pickle
```

```python
with open('MFF-pytorch/ndcg_MFF_ngt_300_RGBFlow_efficientnetv2_segment4_3f1c_ngt_300_4mff3f1c_do50_epochs_160_fr3_3_best.pkl', 'rb') as f:
    ndcg = pickle.load(f)
```

```python
df = pd.DataFrame.from_dict(ndcg)
```

```python
def plot_ndcg(K):
    baselines = {
        1:  0.403,
        5:  0.438,
        10: 0.456,
        20: 0.480,
    }
    
    plt.figure(figsize=(15,8))
    df[f'ndcg@{K}'].plot(kind='hist', bins=10)
    plt.title(f'NDGC@{K} Ditribution', size=24)
    plt.xlabel('NDCG', size=18)
    plt.ylabel('Frequency', size=18)
    plt.xticks(size=16)
    plt.yticks(size=16)
    # Baseline
    y_min, y_max = plt.gca().get_ylim()
    plt.vlines(baselines[K], y_min, y_max * 0.90, color='red', label='baseline')
    plt.text(baselines[K], y_max * 0.91, 'Baseline', ha='center', size=16)
    plt.show()

for K in [1,5,10,20]:
    plot_ndcg(K)
```
