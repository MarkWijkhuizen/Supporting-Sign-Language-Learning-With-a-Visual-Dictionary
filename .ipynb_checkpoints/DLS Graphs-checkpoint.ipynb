```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from tqdm.notebook import tqdm

import math
```

```python
df = pd.DataFrame(columns=['Epsilon', 'Values'])
```

```python
fig = plt.figure(figsize=(15,8))
fig.set_facecolor('white')
fig.set_edgecolor('white')

for eps in [0.100, 0.050, 0.025, 0.010]:
    plt.plot([1 / max(eps, 1-(r/100)) for r in range(0, 101)], label=eps, linewidth=5)
     
plt.grid()
plt.xlabel('Relevance', size=24, labelpad=5)
plt.ylabel('Loss Scale', size=24, labelpad=5)
plt.xticks(size=18)
plt.yticks(size=18)
plt.xticks(np.arange(0, 101, 10), np.arange(0, 1.1, 0.1).round(1), size=16)
plt.yticks(size=16) 
plt.title(f'Dynamic loss scales for different epsilon values', size=32, pad=10)
title = plt.legend(prop={'size': 24}, title='Epsilon').get_title()
title.set_fontsize(24)
plt.plot()
```
